# using GridSearch and Cross validation

# By default in logistic regression the default parameters are:
- C: 1.0,
- penalty: L2
- solver: lfgs
i will change the default parameters and try many using grid search 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'penalty': ['l1','l2','elasticnet'],
    'C': [0.1, 0.01, 5, 10, 50, 100],
    'solver': ['saga', 'liblinear', 'lbfgs','newton-cg'], 
    'l1_ratio': [0, 0.05, 0.01, 0.1 , 1]
}
```

```python
clf = GridSearchCV(model, cv = 3 param_grid= param_grid)
```

```python
clf.fit(X_train_vectorized, y_train)
```

```python
clf.best_params_
```

```python
y_hat = clf.predict(X_test_vectorized)

score = accuracy_score(y_test, y_hat)
print("The accuracy score after applying grid search cv is", score)
```
